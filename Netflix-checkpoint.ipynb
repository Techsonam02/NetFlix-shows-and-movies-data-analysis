**Genre Analysis**

```python
df_genres=df.assign(genres=df['genres'].str.split(',')).explode('genres')
```

```python
genres_counts=df_genres['genres'].value_counts()
```

```python
genres_counts
```

```python
top_genres=genres_counts.head(13)
plt.figure(figsize=(25,14))
ax=sns.barplot(x=top_genres.index,y=top_genres.values,palette='viridis')
for bars in ax.containers:
    ax.bar_label(bars)
    
plt.title('Distribution of genres')
plt.xlabel('Genres')
plt.ylabel('No of titles')
plt.show()
```

***We can say that the majority of movies belong to the international genre. Additionally , an equal no. of movies were created in the action-adventure and dramas genres.***
